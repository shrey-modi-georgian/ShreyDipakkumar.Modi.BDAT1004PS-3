```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('arcade-revenue-vs-cs-doctorates.csv')

df.head(), df.columns
```

```python
plt.figure(figsize=(10, 6))
scatter = plt.scatter(df['Computer Science Doctorates Awarded (US)'],
                      df['Total Arcade Revenue (billions)'],
                      c=df['Year'],
                      cmap='viridis',
                      s=100)

plt.xlabel('Computer Science Doctorates Awarded (US)')
plt.ylabel('Total Arcade Revenue (billions)')
plt.title('Relationship Between Arcade Revenue and CS PhDs Awarded (2000-2009)')

cbar = plt.colorbar(scatter)
cbar.set_label('Year')

plt.show()
```
