```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('actor_kill_counts.csv')

df.head(), df.columns
```

```python
# Sorting the data by 'Count' in ascending order
sorted_kill_counts = df.sort_values(by='Count')

plt.figure(figsize=(10, 8))
bars = plt.barh(sorted_kill_counts['Actor'], sorted_kill_counts['Count'], color='skyblue')
plt.xlabel('Kill Count')
plt.ylabel('Actor')
plt.title('Deadliest Actors in Hollywood Sorted by Kill Count')

for bar in bars:
    plt.text(bar.get_width(), bar.get_y() + bar.get_height()/2,
             f'{int(bar.get_width())}',
             va='center')

plt.show()
```
