```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('us-marriages-divorces-1867-2014.csv')

df.head(), df.columns
```

```python
# Selecting years 1900, 1950, and 2000
selected_years = [1900, 1950, 2000]
data_filtered = df[df['Year'].isin(selected_years)]

plt.figure(figsize=(10, 6))
width = 0.35 
index = np.arange(len(selected_years))

bar1 = plt.bar(index - width/2, data_filtered['Marriages_per_1000'], width, label='Marriages per 1000')
bar2 = plt.bar(index + width/2, data_filtered['Divorces_per_1000'], width, label='Divorces per 1000')


plt.xlabel('Year')
plt.ylabel('Count per 1000 People')
plt.title('Comparison of Marriages and Divorces per Capita in the U.S. for 1900, 1950, and 2000')
plt.xticks(index, selected_years)
plt.legend()

plt.show()
```
