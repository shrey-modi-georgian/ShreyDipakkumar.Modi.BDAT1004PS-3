```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('roman-emperor-reigns.csv')

df.head(), df.columns
```

```python
# Categorizing the 'Cause_of_Death' to distinguish between 'Assassinated' and 'Not Assassinated'
df['Assassinated'] = df['Cause_of_Death'].apply(lambda x: 'Assassinated' if 'Assassinated' in x else 'Not Assassinated')

# To count the values 
assassination_counts = df['Assassinated'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(assassination_counts, labels=assassination_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Fraction of Roman Emperors Assassinated')
plt.axis('equal')
plt.show()
```
